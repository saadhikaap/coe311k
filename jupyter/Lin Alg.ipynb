```python
import numpy as np
```

```python
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

print(v1)
print(v2)
```

```python
v_add = v1+v2
v_sub = v1-v2

print(v_add)
print(v_sub)
```

```python
scale_v1 = v1*2
print(scale_v1)
```

```python
print(np.dot(v1,v2))
```

```python
#magnitude 
mag = np.linalg.norm(v1)
print(mag)
```
