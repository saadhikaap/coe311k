```python
import math
import random 

def approxpi(numS):
    c = 0
    for i in range(numS):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if (x**2 + y**2 < 1):
            c+=1
    pi = (4 * c)/numS
    return pi
        

print(approxpi(50))
print(approxpi(80))
print(approxpi(500))
```
