```python
import math
import random 
import matplotlib.pyplot as plt 
```

```python


#this is my approximation function, it takes a value and runs the loop that many times
#input defines the number of data points used for the approximation 

def approxpi(numS): 
    c = 0
    for i in range(numS):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if (x**2 + y**2 < 1):
            c+=1
    pi = (4 * c)/numS
    return pi

#this is accuracy function, it defines the test pi, and calcs the difference between the calculated pi and real pi 
# the input is the value that approxpi returns 
def accuracy(calc):
    real_pi = math.pi
    test_pi = f"{real_pi:15}"
    test_pi = float(test_pi)
    error_val = abs(test_pi-calc)
    return error_val
```

```python
x_steps = [] #this list will collect the iterated values to be the x axis in the plot 
error = [] # this list will hold the error values and be the y axis 
for i in range (10,10000,20): 
    x_steps.append(i)
    error.append(accuracy(approxpi(i))) 
    
#in the function, i run the approximation, i directly have that input into accuracy and append that answer to the error list
```

```python
plt.plot(x_steps, error, 'r-')
plt.show()
```
