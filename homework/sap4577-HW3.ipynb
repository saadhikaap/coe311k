use for you approximation

```python
import math
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def taylor_sin(n, x): #n = number of terms, x = what value you find sin of 
    num = 0
    den = 0
    ans = 0
    for i in range(n):
        num = x**(2*i+1)
        den = math.factorial(2*i+1)
        ans = ans + ((-1)**i)*(num/den) 
    return ans

def abs_error(calc, x):
    true = np.sin(x)
    return abs(calc-true)
```

```python
#task: find sin x for 0-10 using 1,2,3,4 terms. 
#I want to make 10 lists, each with 4 data points, and plot them on the same plot to compare error
errors = [[] for _ in range(101)]  #makes a list of 10 lists 
x_val = [i for i in range(1, 5)]

x=0
c=0
while x <= 10: #i = x value to solve 
    for j in range(1,5):#j = n number of terms
        errors[c].append(abs_error(taylor_sin(j, x), x))
    x += 0.1
    c += 1
```

```python
for i in range(101):
    plt.plot(x_val, errors[i])
```

```python
comparison = []
for i in range(1,5):
    comparison.append(abs_error(taylor_sin(i, 12), 12))

#comparing using 1-4 terms for x=10 and x=12 
plt.plot(x_val, errors[9],
         x_val, comparison)
```

```python
terms10_7, terms10_15 = [], [] 

x2_val = [i for i in range(1,11)]

for i in range(1,11):
    terms10_7.append(abs_error(taylor_sin(i, 7), 7))
    terms10_15.append(abs_error(taylor_sin(i, 15), 15))
    
#comparing using 1-10 terms for x=7 and x=15 
plt.subplot(1, 2, 1)
plt.plot(x2_val, terms10_7,)
plt.subplot(1, 2, 2)
plt.plot(x2_val, terms10_15, 'b-')
```

Discuss how the truncation error behaves as you include more terms. Does the error decrease uniformly for all values of x? Does the truncation error behave differently for small vs. large values of x (try feeding larger values of x)? (a few sentences is fine)

As you include more terms, the truncation error gets closer to converging. The higher the x value, the quicker the values of the first few terms grow. This means that the higher the error, the more terms needed to get an accurate response. As can be seen in the two graphs above, the error is significantly larger on the right graph for x=15 than for x=7. Likewise, x=7 converged within 10 terms and the error dropped back down, but the right graph for x=15 is still in the process of converging with 10 terms, you need more terms to get an accurate answer. 

Ultimately, the smaller the value of x is, the fewer terms needed in the truncated series to arrive at a good approximation of sin(x). The larger a value of x is, the more terms you need to arrive at a simmilarly accurate approximation.
