Code up Taylor Series for sin(x) in Python. Your Taylor's algorithm needs to have the capability of varying how many terms you're going to
use for you approximation

```python
import math
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def taylor_sin(n, x): #n = number of terms, x = what value you find sin of 
    num = 0
    den = 0
    ans = 0
    for i in range(n):
        num = x**(2*i+1)
        den = math.factorial(2*i+1)
        ans = ans + ((-1)**i)*(num/den) 
    return ans

def abs_error(calc, x):
    true = np.sin(x)
    return abs(calc-true)
```

```python
#task: find sin x for 0-10 using 1,2,3,4 terms. 
#I want to make 10 lists, each with 4 data points, and plot them on the same plot to compare error
errors = [[] for _ in range(11)]  #makes a list of 10 lists 
x_val = [i for i in range(1,5)]

for i in range(11): #i = x value to solve 
    for j in range(1,5):#j = n number of terms
        errors[i].append(abs_error(taylor_sin(j, i), i))
```

```python
plt.plot(x_val, errors[0],        
        x_val, errors[1],
        x_val, errors[2],
        x_val, errors[3],
        x_val, errors[4],
        x_val, errors[5],
        x_val, errors[6],
        x_val, errors[7],
        x_val, errors[8],
        x_val, errors[9],
        x_val, errors[10])
```

```python
comparison = []
for i in range(1,5):
    comparison.append(abs_error(taylor_sin(4, i), i))
   
print(x_val)
print(comparison)
print(errors[5])

plt.plot(x_val, errors[3],
         x_val, comparison)
```
