Code up Taylor Series for sin(x) in Python. Your Taylor's algorithm needs to have the capability of varying how many terms you're going to
use for you approximation

```python
import math
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def taylor_sin(n, x): #n = number of terms, x = what value you find sin of 
    num = 0
    den = 0
    ans = 0
    for i in range(n):
        num = x**(2*i+1)
        den = math.factorial(2*i+1)
        ans = ans + ((-1)**i)*(num/den) 
    return ans

def abs_error(calc, x):
    true = np.sin(x)
    return abs(calc-true)
```

```python
#task: find sin x for 0-10 using 1,2,3,4 terms. 
#I want to make 10 lists, each with 4 data points, and plot them on the same plot to compare error
errors = [[] for _ in range(10)]  #makes a list of 10 lists 
x_val = [i for i in range(1,5)]

for i in range(10): #i = x value to solve 
    for j in range(1,5):#j = n number of terms
        errors[i].append(abs_error(taylor_sin(j, i), i))

print(errors)
```
