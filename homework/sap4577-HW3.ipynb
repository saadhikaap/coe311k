Code up Taylor Series for sin(x) in Python. Your Taylor's algorithm needs to have the capability of varying how many terms you're going to
use for you approximation

```python
import math
```

```python
def taylor_sin(n, x):
    num = 0
    den = 0
    sum = 0
    for i in range(n):
        num = x**(2*n+1)
        den = math.factorial(2*n+1)
        sum = sum + ((-1)**n)*(num/den)
    return sum
        
```

```python
print(taylor_sin(10,1)) 
    
```
