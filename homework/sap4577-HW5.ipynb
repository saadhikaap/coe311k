```python
import numpy as np 
import random
import math 
```

```python
def generate_vals(m, b):
    x = np.array([])
    y = np.array([])
    for i in np.arange(0.0, 1.01, 0.01):
        x = np.append(x, i)
        val = (m*i)+b + random.random()
        y = np.append(y, val)
    return x, y

def manual_solve(x, y):
    y_avg = np.mean(y)
    x_avg = np.mean(x)
    sum_x, sum_tot = 0, 0
    for i in range(len(x)):
        sum_x += (i-x_avg)**2
        sum_tot += (x[i]-x_avg)*(y[i]-y_avg)
    m = sum_tot/sum_x
    b = y_avg - (m*x_avg)
    return m, b

def auto_solve(x, y):
    A = np.vstack([x, np.ones(len(x))]).T
    m, b = np.linalg.lstsq(A, y, rcond=None)[0]
    return m, b
```

```python
x, y = generate_vals(2.1, 5.4)
m_man, b_man = manual_solve(x, y)
m_auto, b_auto = auto_solve(x, y)

print(f"Manual Solver: Slope: {m_man}, Intercept: {b_man}")
print(f"Automatic Solver: Slope: {m_auto}, Intercept: {b_auto}")
```
