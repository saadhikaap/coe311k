```python
import numpy as np
import matplotlib.pyplot as plt

# Exact solution
def exact(x):
    return np.exp(-2*np.sin(x))

# Step sizes
h_values = [0.01, 0.05, 0.15, 0.2, 0.25, 0.5, 0.75, 1.0]

avg_errors = []
local_errors = []

x_target = 15  # point near 15

```

```python
plt.figure()

for h in h_values:
    x = np.arange(0, 10*np.pi, h)
    N = len(x)
    
    y = np.zeros(N)
    y[0] = 1
    
    # Euler method
    for n in range(1, N):
        y[n] = y[n-1] + h * (-2*y[n-1]*np.cos(x[n-1]))
    
    # Exact solution
    y_exact = exact(x)
    
    # Average error
    error = np.abs(y - y_exact)
    avg_errors.append(np.mean(error))
    
    # Local error near x ≈ 15
    idx = np.argmin(np.abs(x - x_target))
    local_errors.append(error[idx])
    
    # Plot Euler solutions
    plt.plot(x, y, label=f'h={h}')
```

```python
# Plot exact solution
x_fine = np.linspace(0, 10*np.pi, 1000)
plt.plot(x_fine, exact(x_fine), 'k--', label='Exact')

plt.title("Euler Method for Different h")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# Plot average error vs h
plt.figure()
plt.plot(h_values, avg_errors, 'o-')
plt.xlabel("h")
plt.ylabel("Average Error")
plt.title("Average Error vs Step Size")
plt.show()
```

```python

# Plot local error near x ≈ 15
plt.figure()
plt.plot(h_values, local_errors, 'o-')
plt.xlabel("h")
plt.ylabel("Local Error at x ≈ 15")
plt.title("Local Error vs Step Size")
plt.show()
```

Euler’s method was used to approximate the solution of (dy/dx = -2y\cos(x)) on ([0, 10\pi]) for various step sizes (h), 
and results were compared to the exact solution (y = e^{-2\sin(x)}). Smaller step sizes closely match the exact solution, while larger step sizes show deviation. The average error increases almost linearly with (h), consistent with Euler’s first-order accuracy (O(h)). Similarly, the local error near (x\approx 15) increases as (h) increases due to accumulated numerical error. Overall, the results demonstrate that decreasing the step size significantly improves accuracy, while larger step sizes lead to greater error.
