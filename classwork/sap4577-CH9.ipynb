```python
import numpy as np
import random 
```

```python
X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])
# Stack the design matrix with a column of ones (for the intercept)
A = np.vstack([X, np.ones(len(X))]).T
# Solve using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Slope: {m}, Intercept: {c}")
```

```python
# Part 1
X_rand = np.array([])
Y_rand = np.array([])

while np.size(X_rand) < 5:
    temp = random.randint(1,10)
    if temp in X_rand: 
        continue
    else: 
        X_rand = np.append(X_rand, temp)
        
for i in range(5):
    Y_rand = np.append(Y_rand, 10*random.random())

X_sort = np.sort(X_rand)
print(X_sort)
print(Y_rand)
```

```python
A1 = np.vstack([X_sort, np.ones(len(X_sort))]).T
m_new, c_new = np.linalg.lstsq(A1, Y_rand, rcond=None)[0]
print(f"Slope: {m_new}, Intercept: {c_new}")
```

```python
# Part 2 
# y = 1.6x + 24.1 

m_calc = 1.6
b_calc = 24.1 

X_test = np.array([])
Y_test = np.array([])

for i in np.arange(0.0, 1.01, 0.01):
    X_test = np.append(X_test, i)
    val = (m_calc*i)+b_calc + random.random()
    Y_test = np.append(Y_test, val)
    

A2 = np.vstack([X_test, np.ones(len(X_test))]).T
m_ans, b_ans = np.linalg.lstsq(A2, Y_test, rcond=None)[0]

print(f"Actual Slope: {m_calc}, Actual Intercept: {b_calc}")
print(f"Calculated Slope: {m_ans}, Calculated Intercept: {b_ans}")
```
