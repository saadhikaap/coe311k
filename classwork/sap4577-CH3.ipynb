```python
import math
import random 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def approxpi(numS):
    c = 0
    for i in range(numS):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if (x**2 + y**2 < 1):
            c+=1
    pi = (4 * c)/numS
    return pi
```

```python
real_pi = np.pi
calc_pi = approxpi(1000)
x = []
abs_error=[] # |x_true - x_calc|
rel_error=[] # | (x_true - x_calc)/x_true |

for i in range(100):
    x.append(i)
    calc_pi = approxpi(1000)
    abs_error.append(abs(real_pi-calc_pi))
    rel_error.append(abs((real_pi-calc_pi)/real_pi))
```

```python
plt.plot(x,abs_error,'r-',
        x, rel_error,'b-')
plt.show()
```
