```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x = np.linspace(0, 2*np.pi, 50)
x2 = np.linspace(0,50,50)
circ_x = np.linspace(-1, 1, 50)
circ_y = np.sqrt(1 - circ_x**2)

plt.subplot(2, 2, 1) 
plt.plot(x, np.sin(x), 'r')

plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x), 'g')

plt.subplot(2,2,3)
plt.plot(x2, x2*2, 'm',
        x2, x2*-1, 'b')

plt.subplot(2,2,4)
plt.plot(circ_x, circ_y, 'k',
        circ_x, -circ_y, 'k')


plt.show()
```
