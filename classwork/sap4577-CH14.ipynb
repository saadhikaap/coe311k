```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and the domain
f = lambda x: np.sin(x)
a, b = 0, 2 * np.pi
N = 50

# Create a uniform grid
x = np.linspace(a, b, N)
y = f(x)

# Step size
h = (b - a) / (N - 1)

# Exact derivative
exact = np.cos(x)

# Central Difference
df_central = (f(x + h) - f(x - h)) / (2 * h)

# Forward Difference
df_forward = (f(x + h) - f(x)) / h

# Backward Difference
df_backward = (f(x) - f(x - h)) / h
```

```python
# Create 4 subplots
plt.figure(figsize=(12, 8))

# 1. Exact derivative
plt.subplot(2, 2, 1)
plt.plot(x, exact, label='Exact (cos(x))')
plt.title('Exact Derivative')
plt.legend()

# 2. Forward Difference
plt.subplot(2, 2, 2)
plt.plot(x, exact, label='Exact')
plt.plot(x, df_forward, 'o--', label='Forward Diff')
plt.title('Forward Difference')
plt.legend()

# 3. Backward Difference
plt.subplot(2, 2, 3)
plt.plot(x, exact, label='Exact')
plt.plot(x, df_backward, 'o--', label='Backward Diff')
plt.title('Backward Difference')
plt.legend()

# 4. Central Difference
plt.subplot(2, 2, 4)
plt.plot(x, exact, label='Exact')
plt.plot(x, df_central, 'o--', label='Central Diff')
plt.title('Central Difference')
plt.legend()

plt.tight_layout()
plt.show()
```
