```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
k = 1.0
x_start = 0
x_end = 10
y_0 = 1.0
v_0 = 0.0

# Step sizes to test
h_values = [0.01, 0.02, 0.03, 0.05, 0.1, 0.2, 0.3, 0.5]

plt.figure(figsize=(12, 6))

for h in h_values:
    #Construct Euler coefficient matrix
    A = np.array([
        [1, h],
        [-k * h, 1]
    ])
    #Compute eigenvalues

    eigenvalues = np.linalg.eigvals(A)
    magnitudes = np.abs(eigenvalues)

    # Stability check
 
    print(f"\nStep size h = {h}")
    print(f"Eigenvalues: {eigenvalues}")
    print(f"Magnitudes: {magnitudes}")

    if np.all(magnitudes < 1):
        print("Solution is Stable")
    else:
        print("Solution may not be Stable ;)")

    # Euler Simulation

    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))

    y_values[0] = y_0
    v_values[0] = v_0

    for i in range(1, len(x_values)):
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]

        y_values[i] = y_n + h * v_n
        v_values[i] = v_n + h * (-k * y_n)

    # Plot results
    plt.plot(x_values, y_values, label=f'h = {h}')

# plot formatting
plt.xlabel('x')
plt.ylabel('y')
plt.title("Euler Method Stability vs Step Size")
plt.legend()
plt.grid(True)
plt.show()
```
